```python
#manipulation data
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data.info()
```

```python
#menghitung jumlah nilai kosong disetiap kolom
data.isna().sum()
```

```python
#menghapus colom id dan unnamed
data.drop('id',axis=1,inplace=True)
data.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
data.head()
```

```python
len(data)
```

```python
data.diagnosis.unique()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
#split data
x=data.drop('diagnosis',axis=1)
y=data.diagnosis

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print("Number transactions x_train dataset: ", x_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions x_test dataset: ", x_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
#lakukan oversampling untuk membalance data
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_sample(x_train, y_train.ravel())

print('After OverSampling, the shape of train_X: {}'.format(x_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
#tahap normalisasi
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
scaler = MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model=RandomForestClassifier(n_estimators = 100)
```

```python
RF_model.fit(x_train, y_train)
```

```python
RF_model.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
            #'n_estimators' : (10,30,50,70,90,100)
              'criterion' : ('gini', 'entropy')
             , 'max_depth' : (3,5,7,9,10)
             , 'max_features' : ('auto', 'sqrt')
             , 'min_samples_split' : (2,4,6)
             }
```

```python
RF_grid = GridSearchCV(RandomForestClassifier(n_jobs = -1, n_estimators = 100), param_grid = parameters, cv = 5, verbose = 0)
```

```python
RF_grid_model = RF_grid.fit(x_train, y_train)
```

```python
RF_grid_model.best_params_
```

```python
RF_grid_model.score(x_test, y_test)
```

```python
RFC = RF_grid_model.best_estimator_
estimators_tree_99 = RFC.estimators_[99]

estimators_tree_3 = RFC.estimators_[3]

estimators_tree_47 = RFC.estimators_[47]
```

```python
estimators_tree_3.n_features_
```
